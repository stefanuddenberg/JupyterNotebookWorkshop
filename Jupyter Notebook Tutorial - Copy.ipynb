## Getting set up

1. [Install Anaconda](https://www.anaconda.com/download/)
    - I'd recommend getting the latest version of Python (version 3.6 at time of writing)
    <br><br>
    
2. Install necessary packages with: 
    - `pip install package_name`
    - You might have to preface that with `sudo` if you're on a Mac.
    <br><br>
    
3. You're probably going to want the following packages (though some may already be installed via Anaconda):
    - jupyter
    - jupyter_contrib_nbextensions
        - run the following command for this after install: `jupyter contrib nbextension install --user`
    - jupyterthemes
        - use if you're not happy with the default aesthetics of the notebook
        - run at terminal for (most of) my aesthetic setup: `jt -t grade3 -fs 12 -tfs 12 -nfs 115 -cellw 88% -T`
    - matplotlib
    - numpy
    - pandas
    - prettypandas
    - pymatbridge
        - for using MATLAB
        - step 2: go to *matlabroot*\extern\engines\python and run `python3 setup.py install`
    - rpy2
        - for using R
        - [See here for further instructions](https://github.com/IRkernel/IRkernel)
    - scipy
    - seaborn
    - statsmodels
    - wes (optional package for Wes Anderson-style color palettes)
    <br><br>
    
4. Run Jupyter notebook from terminal or cmd:
    - `jupyter notebook`
    - Make sure to cd into the directory you want to run it in (or at least a directory *higher* than the one you want; you can't go back up from within the notebook instance!)
    <br><br>

#### Emphasis
*Italics* are made by surrounding a word or phrase with asterisks, or with underscores, _like so_.
<br>

**Bold** words are made by surrounding a word or phrase with 2 asterisks on each end.
<br>

**_You can make a phrase both bold and Italic_** by combining the above!
<br><br>

## Data analysis with pandas

Allow plots in the notebook itself, and enable some helpful functions.

````python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina' # high-res graphs
%config InlineBackend.print_figure_kwargs = {'bbox_inches':'tight'} # no extra white space
%config InlineBackend.figure_format = 'svg'
````

<br>
Import example data.

````python
import pandas as pd

data = pd.read_excel('test_data.xlsx') # xls only lets you have 256 cols!
data.head()
````

<br>
Plot histogram grouped by control. Make sure both have the same x and y limits.

````python
data['sample'].hist(by=data['control'], sharex=True, sharey=True)
````

Overlaid histograms.

````python
control = data.groupby('control')

# You can also add several arguments below like bins=20, or normed=True
figure, axes = control['sample'].plot(kind='hist', normed=False, alpha=.5, legend=True) 
axes.legend(['no control', 'control']) # loc=(%x, %y) to change position of legend

plt.show()
````

Plot simple line graph with sample data.

````python
line_data = range(1,10)

plt.figure()
plt.title("Example Graph", size="xx-large")
plt.xlabel("X-Axis Label", size="x-large")
plt.ylabel("Y-Axis Label", size="x-large")
plt.xlim(0,10)
plt.ylim(0,10)
plt.plot(line_data, 'b*-', markersize=10, linewidth=3, label='Sample Data') # b*- means blue star marker with line
plt.tick_params(axis="both", which="major", labelsize=14)
plt.legend(loc=(0.25, 0.75), scatterpoints=1)
plt.show()
````

## Run R code

Note that this requires running from a Python 3 instance of Jupyter (in my case, at least).

````python
import pandas as pd
df = pd.DataFrame({'Letter': ['a', 'a', 'a', 'b','b', 'b', 'c', 'c','c'],
                   'X': [4, 3, 5, 2, 1, 7, 7, 5, 9],
                   'Y': [0, 4, 3, 6, 7, 10, 11, 9, 13],
                   'Z': [1, 2, 3, 1, 2, 3, 1, 2, 3]})
````

<br>
Load extension allowing one to run R code from within a Python notebook.

````python
%load_ext rpy2.ipython
````

<br>
Do stuff in R with cell or line magics.

````python
%%R -i df
library("ggplot2")
ggplot(data = df) + geom_point(aes(x = X, y = Y, color = Letter, size = Z))
````
